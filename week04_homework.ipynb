```python
#2번
college = pd.read_csv('./data/data/college.csv')
college.head()
```

```python
college2 = college[['INSTNM','CITY','UGDS']]
college2.head()
```

```python
college2.nlargest(100,'UGDS').head()
```

```python
(
    college[['INSTNM','CITY','UGDS']]
    .sort_values('UGDS',ascending=False)
)
```

```python
(
    college[['INSTNM','CITY','UGDS']]
    .sort_values('UGDS',ascending=False)
    .drop_duplicates(subset='UGDS')
)
```

```python
college3 = college[['INSTNM','STABBR','RELAFFIL','UGDS']]
college3.head()
```

```python
college[['INSTNM','STABBR','RELAFFIL','UGDS']].groupby('RELAFFIL',as_index = False).apply(lambda df : df.sort_values(
     'UGDS', ascending = False
         ).head(1)
 ).droplevel(0).sort_values('RELAFFIL',ascending=False)
```

```python
college[['INSTNM','STABBR','RELAFFIL','UGDS']].groupby('STABBR',as_index = False).apply(lambda df : df.sort_values(
     'UGDS', ascending = False
         ).head(1)
 ).droplevel(0).sort_values('RELAFFIL',ascending=False)
```

```python
college[['INSTNM','STABBR','RELAFFIL','UGDS']].groupby('INSTNM',as_index = False).apply(lambda df : df.sort_values(
     'UGDS', ascending = False
         ).head(1)
 ).droplevel(0).sort_values('RELAFFIL',ascending=False)
```
